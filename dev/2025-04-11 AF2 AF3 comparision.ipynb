# Comparison between AF2 and AF3

```python
# Imports
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from pathlib import Path
import seaborn as sns

class bcolors:
    HEADER = '\033[95m'
    OKBLUE = '\033[94m'
    OKCYAN = '\033[96m'
    OKGREEN = '\033[92m'
    WARNING = '\033[93m'
    FAIL = '\033[91m'
    ENDC = '\033[0m'
    BOLD = '\033[1m'
    UNDERLINE = '\033[4m'
```

```python
# Settings

# Path to resource folder with the structures and metadata tables
path_resources = Path(r"D:\Eigene Datein\dev\Uni\JGU Bio Bachelorthesis\Daten\resources")

path_AF2 = path_resources / "AF2"
path_AF3 = path_resources / "AF3"
path_solved = path_resources / "solved"
```

```python
dataAF2 = pd.read_csv(path_AF2 / "AF2_metrics.tsv", sep="\t")
# Old metrics
# dataAF2 = pd.read_csv(path_AF2 / "John metrics" / "AF_metrics_all_structures.tsv", sep="\t")
for c in ["chainA_start", "chainA_end", "chainB_start", "chainB_end", "num_mutations", "num_align_atoms_domain", "num_align_resi_domain", "hbonds", "salt_bridges", "hydrophobic_interactions"]:
    if c not in dataAF2.columns:
        print(f"Column {bcolors.FAIL}{c}{bcolors.ENDC} not (yet) in data frame")
        continue
    dataAF2[c] = dataAF2[c].astype(pd.Int64Dtype())
display(dataAF2)
```

```python
dataAF3 = pd.read_csv(path_AF3 / "AF3_metrics.tsv", sep="\t")
for c in ["chainA_start", "chainA_end", "chainB_start", "chainB_end", "num_mutations", "num_align_atoms_domain", "num_align_resi_domain", "hbonds", "salt_bridges", "hydrophobic_interactions"]:
    if c not in dataAF3.columns:
        print(f"Column {bcolors.FAIL}{c}{bcolors.ENDC} not (yet) in data frame")
        continue
    dataAF3[c] = dataAF3[c].astype(pd.Int64Dtype())
display(dataAF3)
```

```python
dataAF = pd.merge(
    left=dataAF3,
    right=dataAF2,
    left_on=["benchmark_set", "prediction_name", "model_id"],
    right_on=["benchmark_set", "prediction_name", "model_id"],
    suffixes=["_AF3", "_AF2"],
    how="inner"
)

display(dataAF)
```

### RMSD

```python
dataAF["delta_RMSD_domain"] = dataAF.apply(lambda r: r["RMSD_domain_AF3"] - r["RMSD_domain_AF2"], axis=1)
dataAF["delta_RMSD_all_atom_peptide"] = dataAF.apply(lambda r: r["RMSD_all_atom_peptide_AF3"] - r["RMSD_all_atom_peptide_AF2"], axis=1)
```

```python
compare_columns = ["RMSD_domain", "RMSD_backbone_peptide", "RMSD_all_atom_peptide", "DockQ"]
RMSD_AF2 = {c:[] for c in compare_columns}
RMSD_AF3 = {c:[] for c in compare_columns}
for i, row_new in dataAF3.iterrows():
    prediction_name = row_new["prediction_name"]
    model_id = row_new["model_id"]

    if model_id != "ranked_0": continue

    if len(row_old := dataAF2[np.logical_and(dataAF2["prediction_name"].str.lower() == prediction_name.lower(), dataAF2["model_id"] == model_id)]) == 0:
        print(f"\t{bcolors.FAIL}{prediction_name} {model_id} does not exist{bcolors.ENDC}")
        continue

    for c in compare_columns:
        RMSD_AF3[c].append(row_new[c])
        RMSD_AF2[c].append(row_old[c].item())
```

```python
for c in compare_columns:
    x, y = RMSD_AF2[c], RMSD_AF3[c]
    ratio = np.array(x)/np.array(y)
    plt.scatter(x, y, alpha=1, s=1)
    plt.plot([0, np.nanmax([x, y])], [0, np.nanmax([x, y])], color='navy', lw=1, linestyle='--')
    plt.xlabel("AF2")
    plt.ylabel("AF3")
    if "RMSD" in c:
        plt.xscale("log")
        plt.yscale("log")
    plt.title(c)
    plt.show()
    
    
    plt.violinplot([ratio[np.isfinite(ratio)]])
    plt.title(c)
    plt.ylabel("ratio AF2/AF3")
    plt.show()
```
